```python
import scipy.special as sp
import numpy as np
from matplotlib import pyplot as plt 
plt.style.use('fivethirtyeight')
```

```python
print(sp.poch(3,2))
sp.factorial(2)
```

```python
def Gauss(a,b,g,x):
    res = 0
    for k in range(50):
        a_k = sp.poch(a,k)
        b_k = sp.poch(b,k)
        g_k = sp.poch(g,k)
        res += a_k*b_k/g_k*x**k/sp.factorial(k)
    return res
```

```python
nus = [1,1.5,2,3,3.5,6]

for nu in nus:
    X = np.linspace(-1,1,100)
    Y = np.zeros(len(X))

    a = -nu 
    b = nu+1
    g = 1

    for i in range(len(Y)):
        x = X[i]
        Y[i] = Gauss(a,b,g,1/2*(1-x))
    plt.plot(X,Y, label='$\\nu = {}$'.format(nu))
plt.legend()
plt.savefig('images/p2_p.png', bbox_inches='tight')
plt.show()
```

```python
nus = [1,1.5,2,3,3.5,6]
for nu in nus:
    X = np.linspace(2,3,100)
    Y = np.zeros(len(X))

    a = (nu+1)/2
    b = (nu+2)/2
    g = nu+3/2

    for i in range(len(Y)):
        x = X[i]
        #np.sqrt(np.pi)/2**(nu+1)*sp.gamma(nu)/sp.gamma(nu+3/2)
        Y[i] = sp.gamma(nu+1)*1/x**(nu+1)*Gauss(a,b,g,1/x**2)
    plt.plot(X,Y)

plt.show()
```
