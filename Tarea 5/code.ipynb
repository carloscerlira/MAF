```python
import numpy as np 
import pandas as pd 
from scipy import special
import matplotlib.pyplot as plt 
from numpy import random
plt.style.use('fivethirtyeight')
```

```python
def f(x):
    if(10<=x and x<20):
        return 0.025
    if(20<=x and x<30):
        return 0.050
    if(30<=x and x<40):
        return 0.025
    return 0 
```

```python
X = np.linspace(0,50,100)
Y = np.zeros(len(X))
for i in range(len(Y)):
    Y[i] = f(X[i])

plt.plot(X,Y,color='royalblue',label='$f(x)$')
plt.legend()
plt.savefig('images/p4_density')
```

```python
def discrete_graph(pmf, X, name):
    Y = np.zeros(len(X))
    for i in range(len(X)):
        Y[i] = pmf(X[i])
    Y = Y

    plt.bar(X, Y, color='royalblue',width=1.0, edgecolor='black',  label=name+' distribution')
    plt.xlabel('$x$',fontsize=12)
    plt.ylabel('$P(X=x)$',fontsize=12)
    plt.legend()
    plt.savefig('images/p7_'+name)
    plt.show()
```

```python
def continous_graph(pdf, X, name, col='royalblue'):
    Y = np.zeros(len(X))
    for i in range(len(X)):
        Y[i] = pdf(X[i])

    plt.plot(X, Y, color=col, label=name+' distribution')
    plt.xlabel('$x$',fontsize=12)
    plt.ylabel('$f_X(x)$',fontsize=12)
    plt.legend()
    plt.savefig('images/p7_'+name)
```

```python
p = 0.2
def pmf(n):
    return (1-p)**(n-1)*p 
n=15
X = np.arange(1,n)
discrete_graph(pmf, X, 'geometric')
```

```python
p=0.3
r=3
def pmf(n):
    return special.binom(n-1, r-1)*p**r*(1-p)**(n-r)
n=30
X = np.arange(r,n)
discrete_graph(pmf, X, 'binomial')
```

```python
a=5
l=1
def pdf(x):
    return l*np.exp(-l*x)*(l*x)**(a-1)/special.gamma(a) 
n=12
X = np.linspace(0,n,100)
continous_graph(pdf, X, 'gamma')
```

```python
def pdf(x):
    return 1/np.pi*1/(1+x**2)
n=10
X = np.linspace(-n,n,100)
continous_graph(pdf, X, 'cauchy')
```

```python
a,b=75,5
def pdf(x):
    return 1/special.beta(a,b)*x**(a-1)*(1-x)**(b-1)
X = np.linspace(0,1,100)
continous_graph(pdf, X, 'beta')

a,b=9,1
X = np.linspace(0,1,100)
continous_graph(pdf, X, 'beta', 'purple')
```
